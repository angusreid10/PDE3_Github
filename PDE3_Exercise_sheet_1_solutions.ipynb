__Q1__

Consider the function $u(x,y) = x^4 - 6x^2y^2 + y^4$.

a) Verify that $u(x,y)$ satisfies the Laplace equation.

b) Use Python to generate a surface plot of $u(x,y)$ in the square given by $0\le x\le1$ and $0\le y\le 1$.

c) Give the boundary conditions fulfilled by $u(x,y)$ on the boundaries of the square given by $0\le x\le1$ and $0\le y\le 1$.
\
\
__Solution__

a) Calculate the first and second partial derivatives of $u(x,y)$ with respect to $x$ and $y$ to get

\begin{align}
\frac{\partial u(x,y)}{\partial x} &= 4x^3 - 12 x y^2 \\
\frac{\partial^2 u(x,y)}{\partial x^2} &= 12x^2 - 12 y^2 & (1.1) \\
\frac{\partial u(x,y)}{\partial y} &= - 12 x^2 y + 4 y^3 \\
\frac{\partial^2 u(x,y)}{\partial y^2} &= - 12x^2 + 12 y^2 & (1.2)
\end{align}

Inserting Eqs. (1.1) and (1.2) into the Laplace equation gives

\begin{align}
\nabla^2 u = \frac{\partial^2 u(x,y)}{\partial x^2} + \frac{\partial^2 u(x,y)}{\partial y^2} = 12x^2 - 12 y^2 - 12x^2 + 12 y^2 = 0
\end{align}

thus showing that $u(x,y)$ satisfies the Laplace equation.


```python
# b)
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator , FormatStrFormatter
import numpy as np
# Solution to the Laplace equation on a square
# Set the x and y grid points and the meshgrid
N = 20 ;
x = np.linspace(0,1,N);
y = np.linspace(0,1,N);
X,Y = np.meshgrid(x,y);
# Define the function and generate the output on the meshgrid
Z = X**4 - (6*X*X * Y*Y) + np.power(Y,4);
# Generate the plot
fig = plt.figure()
fig.set_size_inches(10,10)
ax = plt.axes(projection='3d')
# Plot the surface
surf = ax.plot_surface(X,Y,Z,cmap=cm.coolwarm,linewidth=0,antialiased=False)
# Customize the z axis
ax.set_zlim(-4,1)
ax.zaxis.set_major_locator(LinearLocator(6))
ax.zaxis.set_major_formatter(FormatStrFormatter('%.1f'))
ax.set_xlabel('xaxis')
ax.set_ylabel('yaxis')
ax.set_zlabel('u(x,y)')
# Add a color bar which maps values to colors
fig.colorbar(surf,shrink=0.5,aspect=10,location="left")
plt.show()
```

c) By inserting the boundary values into the equation for $u(x,y)$ we get the following boundary conditions 

\begin{align}
u(x, 0) &= x^4, \qquad &0\le x\le 1 & \to \text{bottom boundary} \\
u(x, 1) &= x^4 - 6x^2 + 1, \qquad &0\le x\le 1 & \to \text{top boundary} \\
u(0, y) &= y^4, \qquad &0\le y\le 1 & \to \text{left boundary} \\
u(1, y) &= 1 - 6y^2 + y^4, \qquad &0\le y\le 1 & \to \text{right boundary} 
\end{align}

__Q2__

Consider the the function $u(x,y) = e^{-x} \sin(y)$.

a) Verify that $u(x,y)$ satisfies the Laplace equation.

b) Use Python to generate a surface plot of $u(x,y)$ in the square given by $0\le x\le1$ and $0\le y\le \pi$.

c) Give the boundary conditions fulfilled by $u(x,y)$ on the boundaries of the square given by $0\le x\le1$ and $0\le y\le \pi$.
\
\
__Solution__

a) Calculate the first and second partial derivatives of $u(x,y)$ with respect to $x$ and $y$.

\begin{align}
\frac{\partial u}{\partial x} &= -e^{-x} \sin(y) \\
\frac{\partial^2 u}{\partial x^2} &= e^{-x} \sin(y) & (2.1) \\
\frac{\partial u}{\partial y} &= e^{-x} \cos(y) \\
\frac{\partial^2 u}{\partial y^2} &= -e^{-x} \sin(y) & (2.2)
\end{align}

Inserting Eqs. (2.1) and (2.2) into the Laplace equation gives

\begin{align}
\nabla^2 u = \frac{\partial^2 u}{\partial x^2} + \frac{\partial^2 u}{\partial y^2} = e^{-x} \sin(y) - e^{-x} \sin(y) = 0
\end{align}

thus showing that $u(x,y)$ satisfies the Laplace equation.

```python
# b)
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator , FormatStrFormatter
import numpy as np
# Solution to the Laplace equation on a square
# Set the x and y grid points and the meshgrid
N = 50
x = np.linspace(0,1,N)
y = np.linspace(0,np.pi,N)
X,Y = np.meshgrid(x,y)
# Define the function and generate the output on the meshgrid
Z = np.exp(-X) * np.sin(Y)
# Generate the plot
fig = plt.figure()
fig.set_size_inches(10,10)
ax = plt.axes(projection='3d')
# Plot the surface
surf = ax.plot_surface(X,Y,Z,cmap=cm.coolwarm,linewidth=0,antialiased=False)
# Customize the z axis
ax.set_zlim(0,1)
ax.zaxis.set_major_locator(LinearLocator(6))
ax.zaxis.set_major_formatter(FormatStrFormatter('%.01f'))
ax.set_xlabel('xaxis')
ax.set_ylabel('yaxis')
ax.set_zlabel('u(x,y)')
# Add a color bar which maps values to colors
fig.colorbar(surf,shrink=0.5,aspect=8,location="left")
plt.show()
```

c) By inserting the boundary values into the equation for $u(x,y)$ we get the following boundary conditions 

\begin{align}
u(x, 0) &= 0, \qquad &0\le x\le 1 & \to \text{bottom boundary} \\
u(x, \pi) &= 0, \qquad &0\le x\le 1 & \to \text{top boundary} \\
u(0, y) &= \sin(y), \qquad &0\le y\le \pi & \to \text{left boundary} \\
u(1, y) &= e^{-1} \sin(y), \qquad &0\le y\le \pi & \to \text{right boundary} 
\end{align}

__Q3__

Show that the function $u(x,t) = e^{-kt} \cos(mx) \cos(nt)$ is a solution to the partial differential equation

\begin{align}
c^2 \frac{\partial^2 u}{\partial x^2} = \frac{\partial^2 u}{\partial t^2} + 2k \frac{\partial u}{\partial t}
\end{align}

if the constants with parameters $k$, $m$, $n$ and $c$ are related by the equation $n^2 + k^2 = c^2 m^2$.

__Solution__

We start by calculating the partial derivatives of $u(x,t)$ to get

\begin{align}
\frac{\partial u}{\partial t} &= -k e^{-kt} \cos(mx) \cos(nt) - n e^{-kt} \cos(mx) \sin(nt) & (3.1) \\
\frac{\partial^2 u}{\partial t^2} &= k^2 e^{-kt} \cos(mx) \cos(nt) + kn e^{-kt} \cos(mx) \sin(nt) \\
 & \quad + kn e^{-kt} \cos(mx) \sin(nt) - n^2 e^{-kt} \cos(mx) \cos(nt) \\
 &= (k^2 - n^2) e^{-kt} \cos(mx) \cos(nt) + 2kn e^{-kt} \cos(mx) \sin(nt) & (3.2) \\
\frac{\partial u}{\partial x} &= -me^{-kt} \sin(mx) \cos(nt) \\
\frac{\partial^2 u}{\partial x^2} &= -m^2e^{-kt} \cos(mx) \cos(nt) & (3.3) 
\end{align}

Now we substitute the results from equations (3.1)-(3.3) in the partial differential equation. 

\begin{align}
- c^2 m^2e^{-kt} \cos(mx) \cos(nt) &= (k^2 - n^2) e^{-kt} \cos(mx) \cos(nt) + 2kn e^{-kt} \cos(mx) \sin(nt) \\
 & \quad - 2k^2 e^{-kt} \cos(mx) \cos(nt) - 2k n e^{-kt} \cos(mx) \sin(nt) \\
 &= -(k^2 + n^2) e^{-kt} \cos(mx) \cos(nt)
\end{align}

We can divide the resulting expression with $u(x,t)$ to get the required relationship between the four parameters, thus proving that $u(x,t)$ is a solution as long as $c^2 m^2 = k^2 + n^2$ holds.
